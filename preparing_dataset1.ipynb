```python
import pandas as pd
import gzip
```

```python
df1 = pd.read_csv(gzip.open('name.basics.tsv.gz', 'rt'), sep='\t', dtype={4: str})
```

```python
df2 = pd.read_csv(gzip.open('title.basics.tsv.gz', 'rt'), sep='\t')
```

```python
df2['combined'] = df2.apply(lambda row: f"{row['primaryTitle']} ({row['titleType']}) - {row['genres']}", axis=1)
```

```python
df2 = df2.drop(['titleType', 'primaryTitle', 'genres'], axis = 1)
```

```python
df2 = df2.drop(['originalTitle', 'isAdult', 'startYear', 'endYear', 'runtimeMinutes'], axis = 1)
```
